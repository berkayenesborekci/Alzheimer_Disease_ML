# Model Definitions
This notebook defines the three machine learning models used in this project.

```python
# LogisticRegression is a linear classifier from scikit-learn
# RandomForestClassifier is an ensemble of decision trees
# XGBClassifier is a gradient boosting algorithm known for strong performance on tabular data
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

### Model Setup

```python
# Logistic Regression is sensitive to feature scale because it relies on distances
# tree-based models like Random Forest and XGBoost split on thresholds, so scaling has no effect on them
# this set is used during training to decide which input version (scaled or raw) to pass to each model
SCALED_MODELS = {'Logistic Regression'}
```

```python
# Logistic Regression models the probability of a binary outcome using a linear function
# it serves as a simple baseline — if tree-based models don't outperform it, the setup likely has an issue
# max_iter=1000 increases the number of optimization steps to ensure convergence on larger datasets
logistic = LogisticRegression(
    max_iter=1000,
    random_state=42,
)
print('Logistic Regression:', logistic)
```

```python
# Random Forest is an ensemble method that trains multiple decision trees on random subsets of the data
# each tree votes on the prediction and the majority vote is taken as the final output
# n_estimators=200 sets the number of trees — more trees reduce variance but increase training time
# n_jobs=-1 enables parallel training using all available CPU cores
random_forest = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1,
)
print('Random Forest:', random_forest)
```

```python
# XGBoost is a gradient boosting algorithm that builds trees sequentially
# each new tree is trained to minimize the residual errors of the previous trees
# n_estimators=300 sets how many boosting rounds (trees) to run
# learning_rate=0.1 controls the contribution of each tree — smaller values reduce overfitting but need more trees
# max_depth=6 restricts tree depth to prevent the model from memorizing the training data
# eval_metric='logloss' uses log loss as the internal evaluation metric during training
xgboost = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.1,
    eval_metric='logloss',
    random_state=42,
    n_jobs=-1,
)
print('XGBoost:', xgboost)
```

```python
# grouping all models into a dictionary makes it easy to train and evaluate them in a single loop
# this avoids code duplication and makes it straightforward to add or remove models later
models = {
    'Logistic Regression': logistic,
    'Random Forest':       random_forest,
    'XGBoost':             xgboost,
}

print('Models ready:', list(models.keys()))
```
